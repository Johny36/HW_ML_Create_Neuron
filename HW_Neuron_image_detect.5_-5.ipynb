
Single Neuronal Model
Sa presupunem ca avem o matrice de numere (3x3) care arata in felul urmator

m = [
  [5, -1,  0],
  [2,  0, -5],
  [1,  1, -5],
] 

valorile numerice vor fi mereu cuprinse in diapazonul -5 .. 5
Sa se alcatuiasca si sa se antreneze un model cu un singur neuron, care va fi capabil sa detecteze daca matricea contine numere pozitive in prima coloana

modelul trebuie sa dea o valoarea mai mare de 0.7 la iesire atunci cand prima coloana din matrice are TOATE numerele pozitive si valori mai mici ca 0.7 atunci cand pe prima coloana sunt fie numere negative fie zero

deci pentru matricea de mai sus rezultatul neuronului ar fi > 0.7 (ideal 1)




```python
## Batch of Training data

Xtrain = [
    [
    [5, -1,  0],
    [2,  0, -5],
    [1,  1, -5],
    ],
    [
    [ 0, -1,  1],
    [-1,  0,  2],
    [ 1,  1, -5],
    ],
     [
    [ 1,  2, -1],
    [-1,  0,  1],
    [ 2,  0,  5],
    ],
     [
    [-5, -1,  1],
    [ 1,  0,  5],
    [-1,  2, -5],
    ],
    [
    [ 1, 0,  0],
    [-1, 0,  5],
    [ 1,-1, -1]
    ],
    [
    [ 2, 1,  3],
    [ 1, 1, -1],
    [-5, 1, -1]
    ],
    [
    [-1,  1, -5],
    [ 1,  0,  2],
    [ 0, -1, -2]
    ],
]

Ytrain = [
    1,
    0,
    0,
    0,
    0,
    0,
    0,
]

#### Batch of Testing data
Xtest = [
    [ 
    [ 1,  1,  0],
    [ 2,  5,  1],
    [ 1,  1, -5],
    ],
    [ 
    [ 1, -2,  0],
    [ 0,  1, -5],
    [-1,  1, -5],
    ],
    [
    [ 3, -4,  1],
    [ 5,  0, -2],
    [-2,  1, -5],
    ]
]
Ytest = [
    0,
    0,
    0,
]
```

```python
# x - singel values
# X - collections


class LinearNeuronDetect:

    def __init__(self):
        self.weights = [0 for i in range(9)]
        self.bias    = 0

    def sigmoidActivation(self, x):
        e = 2.718281828459045
        return 1/ (1 + e** - x)

    def forward(self, X):
        y = self.bias
        for i in range (9):
            y += X[i] * self.weights[i]

        y = self.sigmoidActivation(y)

        return y

    def __str__(self):

        out = ""
        for i in range(9):
            if i % 3 == 0:
                out += "\n"
            out += f"{self.weights[i]:10.5f} "
            
        out += "\n"       
        out += f"{self.bias:10.5f} "

        return out

def loss(Y, Yp):
    return Y - Yp

def reshapeMatrix(X):
    Xt = []
    for i in range(3):
        for j in range(3):
            Xt.append(   X[i][j]  )

    return Xt
```

### Neuron Training

```python
neuron = LinearNeuronDetect()

lr = 0.01

for epoch in range(1,5000):
    errors = []
    for i in range(7):
        # Forward Pass
        xt = Xtrain[i]
        yt = Ytrain[i]

        xt = reshapeMatrix(xt)
        yp = neuron.forward(xt)

        # Estimate Loss
        e = loss(yt, yp)

        ### BackPropagation / Optimization
        dBias = lr * e
        neuron.bias += dBias

        for j in range(9):
            dWeight  = lr * e * xt[j]
            neuron.weights[j] += dWeight
        errors.append(e)
    ## Epoch Stats
    me = abs(sum(errors) /7) 
    print("-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+")
    print(f"epoch: {epoch}")
    print(f"loss: {me: 10.5f}")
    #print(neuron)


       
```

###Testing

```python
X = Xtrain + Xtest
Y = Ytrain + Ytest
for i in range (10):
    xt = X[i]
    yt = Y[i]
    xt = reshapeMatrix(xt)
    yp = neuron.forward(xt)
    e = loss(yt, yp)
    print(f"loss: {e: 10.5f} Yp{yp: 10.5f}")
     
```
